## Investigation Overview

In this investigation, I wanted to look at the variables that influence drip duration. The main factors found to have an influence were rainy weather, age, and gender proportion.

## Dataset Overview

The data consisted of daily summaries of rideshare and weather data from June 28<sup>th</sup> 2017 to April 30<sup>th</sup> 2019.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe. 
df = pd.read_pickle('daily_summary.pkl')
df['avg_duration_minutes'] = df.avg_duration.astype('timedelta64[s]')/60
df['age'] = df.date.dt.year - pd.to_datetime(df.avg_birth_year, format='%Y').dt.year 
df.male_percentage = df.male_percentage*100
```

## Temperature and Trip Duration on Weekends

This graph scatter plots the average temperatures and average trip duration in minutes on weekends.  
There is also a line of best fit that highlights the general correlation of the data.

This highlights how as Temperature increases, so does the Trip Duration.

```python
plt.figure(figsize=(15,10))
sb.regplot(data=df[df.day =='weekend'], x='TAVG', y='avg_duration_minutes')
plt.xlabel('Daily Average Temperature (Celsius)', fontsize=20)
plt.ylabel('Average Trip Duration (Minutes)', fontsize=20)
plt.title('Temperature vs. Trip Duration on the Weekend', fontsize=30)
plt.ylim(10,18)
plt.xlim(5,30);
```

## Age and Trip Duration on Weekends

This graph scatter plots the average age and average trip duration in minutes on weekends.
There is also a line of best fit that highlights the general correlation of the data.

This highlights how as Age increases, so does the Trip Duration.

```python
plt.figure(figsize=(15,10))
sb.regplot(data=df[df.day =='weekend'], x='age', y='avg_duration_minutes', x_jitter=.5)
plt.xlabel('Average Age', fontsize=20)
plt.ylabel('Average Trip Duration (Minutes)', fontsize=20)
plt.title('Age vs. Trip Duration on the Weekend', fontsize=30)
plt.ylim(10,18)
plt.xlim(32,38);
```

## Percentage of Males and Trip Duration when it is Rainy

This graph scatter plots the percentage of males and average trip duration in minutes on Rainy days.
There is also a line of best fit that highlights the general correlation of the data.

This highlights how as the percentage of males increases, the Trip Duration decreases.

```python
plt.figure(figsize=(15,10))
sb.regplot(data=df[df.rain == 'Rainy'], x='male_percentage', y='avg_duration_minutes')
plt.ylabel('Average Trip Duration (Minutes)', fontsize=20)
plt.xlabel('Percentage of Male Riders', fontsize=20)
plt.title('Percentage of Males vs. Trip Duration when it is Rainy', fontsize=25)
plt.ylim(9,14)
plt.xlim(72,82);
```
